```python
import math

def get_triangular_num(n):
    # returns the nth triangular number
    return int(n*(n+1)/2)

def get_divisors(n):
    # returns a list containing the divisors of n
    divisors = []
    for i in range(1,math.ceil(math.sqrt(n))):
        if n % i == 0:
            divisors.append(i)
            divisors.append(int(n/i))
    return divisors

max_divisors = 0
n = 0
while max_divisors < 500:
    n += 1
    max_divisors = len(get_divisors(get_triangular_num(n)))
get_triangular_num(n)
```
