```python
import math
import itertools

def quad(n, a, b):
    return n**2 + a*n + b

def is_prime(n):
    if n <= 0:
        return False
    for i in range(2, math.ceil(math.sqrt(n))+1):
        if n % i == 0 and n != i:
            return False
    return True

def count_primes(a, b):
    for i in itertools.count():
        if not is_prime(quad(i,a,b)):
            return i

coefficents = max([(a,b) for a in range(-999, 1000) for b in range(-1000,1001)], key = lambda x: count_primes(x[0], x[1]))
print(coefficents[0] * coefficents[1])
```
